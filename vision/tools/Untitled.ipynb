```python
import os
import numpy as np
import json
import matplotlib.pyplot as plt

def load_coco_file(file_path):
    with open(file_path, 'r') as f:
        coco = json.load(f)
    return coco
```

```python
file_path = '/home/fruitspec-lab/FruitSpec/Data/FX_JAI_FSI_V1_COCO/COCO/annotations/instances_train_new.json'
coco = load_coco_file(file_path)
```

```python
ann = coco['annotations']
h = []
w = []
area = []
ratio = []

for a in ann:
    h.append(a['bbox'][2])
    w.append(a['bbox'][3])
    ratio.append(a['bbox'][2] / a['bbox'][3])
    area.append(a['area'])
```

```python
_ = plt.hist(h, bins=50, range=(0, np.max(h)))
plt.title('bbox height dist')
plt.xlabel('Pixels')
plt.ylabel('Count')
```

```python
_ = plt.hist(w, bins=50, range=(0, np.max(w)))
plt.title('bbox width dist')
plt.xlabel('Pixels')
plt.ylabel('Count')
```

```python
_ = plt.hist(ratio, bins=50, range=(0, np.max(ratio)))
plt.title('bbox ratio')
```
